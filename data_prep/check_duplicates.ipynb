```python
import collections
import os 
import glob
root_dir = '/data2/xiangrui_d2/MMC/Dataset/Combine/wav/WATKINS/unsplited'
sub_dir_list = os.listdir(root_dir)
sub_dir_list = [os.path.join(root_dir, sub_dir) for sub_dir in sub_dir_list]
sub_dir_list
```

```python
audio_list = []
for folder_path in sub_dir_list:
    for file_path in glob.glob(folder_path + '/*.wav'):
        audio_list.append(os.path.basename(file_path))
audio_list
```

```python
print([item for item, count in collections.Counter(audio_list).items() if count > 1])
```
