```python
import matplotlib.pyplot as plt
import numpy as np
import librosa
import librosa.display
```

```python
def plot_spec(wav,sr,n_fft=2048,hop_length=512,title = 'Spectrogram'):
    y, sr = librosa.load(wav, sr=sr)
    # D = np.abs(librosa.stft(y))**2
    # S = librosa.feature.melspectrogram(S=y, sr=sr)
    S = librosa.feature.melspectrogram(y=y,n_fft= n_fft, hop_length=hop_length,sr=sr, n_mels=256)
    fig, ax = plt.subplots()
    S_dB = librosa.power_to_db(S, ref=np.max)         
    print(S_dB.shape)
    img = librosa.display.specshow(S_dB[::,100:500], x_axis='time',
                         y_axis='mel', sr=sr,
                         ax=ax)
    fig.colorbar(img, ax=ax, format='%+2.0f dB')
    ax.set(title=title)
```

```python
audio= r'/data2/xiangrui_d2/MMC/Dataset/Combine/wav/Moby/wav/Train/Finback Whale/93-041-0244.ch16.wav'
plot_spec(audio,sr=2000,n_fft=512,hop_length=256,title='Finback Whale')
```
